```python
df.drop_duplicates(inplace=True)
```

```python
df["Age"] = df["Age"].astype(int)
df["Salary"] = df["Salary"].astype(float)
```

```python
df["Department"].value_counts() # count of each departments
```

```python
#basic Stats

print("Mean of Age:", df["Age"].mean())
print("Median of age:", df["Age"].median())
print("Max Salary:", df["Salary"].max())
print("Min Salary:", df["Salary"].min())
print("Mean of Salary:",df["Salary"].mean())
```

```python
ages = df["Age"].to_numpy()
Salaries = df["Salary"].to_numpy()

print("Age STD", np.std(ages))
print("Salary STD:", np.std(Salaries))
print("Salary + 5000 bump: ", Salaries + 5000)
```
