```python
import numpy as np

arr = np.array([10,20,30,40,50])
print(arr)
print(type(arr))
print(arr[0])      # first element
print(arr[1:4])    # from index 1 to 3
print(arr[:3])     # first 3 elements
print(arr[2:])     # from index 2 to end

```

```python
matrix = np.arange(1, 26).reshape(5, 5)
print(matrix)

# Top-left 3x3
print(matrix[:3, :3])
print("*****")
print(matrix[2:,2:])
print("**$**")
print(matrix[1:4, 1:4])
print("**$**")
print(matrix[:, -1])
print("*****")
print(matrix[:, 1])
print("*****")
print(matrix[1, :])
print("*****")
```

```python
arr = np.arange(1,13)
print(arr)

matrix = arr.reshape(3,4)
print(matrix)
```

```python
matrix = np.arange(1, 10).reshape(3, 3)
print(matrix)

print(matrix + 10)
```

```python
weights = np.array([1,2,3])
print(matrix * weights)
```

```python
arr = np.arange(1,101)
matrix = arr.reshape(10, 10)

print(matrix + 5)
```

```python
arr = np.arange(1,101)
matrix = arr.reshape(10, 10)

matrix_even = matrix.copy()

mask = matrix_even % 2 ==0

matrix_even[mask] = matrix_even[mask] * 2

print(matrix_even)
print(matrix)
print(matrix[3:6, 3:6])
```
