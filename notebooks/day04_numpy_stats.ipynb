```python
import numpy as np

arr = np.array([12, 45, 67, 89, 23, 90, 100])

print("Numbers more than 50:", arr[arr > 50])

even_and_gt_50 = arr[(arr % 2 == 0) & (arr > 50)]

print("Extract numbers even AND > 50:", even_and_gt_50)

new_arr = arr.copy()
new_arr[new_arr <= 30] = 0

print("Original:", arr)
print("Modified:", new_arr)
```
