```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
product_A = [200, 250, 180, 300, 320]
product_B = [150, 230, 210, 280, 310]
product_C = [125, 240, 219, 260, 350]

plt.plot(days, product_A, label="Product A", color='green')
plt.plot(days, product_B, label="Product B", marker='o', linestyle='--', color='yellow')
plt.plot(days, product_C, label="Product C", color='orange')

plt.title("Sales Comparison")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()

plt.savefig("sales_comparison.png")   
plt.show()
```

```python
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
df.head()
df.info()
df.describe()
```

```python
df.groupby("sex")["survived"].mean().plot(kind="bar")

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.groupby("pclass")["survived"].mean().plot(kind="bar")

plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Class (1 = First, 3 = Third)")
plt.show()
```

```python
import matplotlib.pyplot as plt

survival_rate = df.groupby(["pclass", "sex"])["survived"].mean().unstack()

survival_rate.plot(kind="bar")

plt.title("Survival Rate by Class and Gender")
plt.ylabel("Survival Rate")
plt.savefig("Survival_Rate.png")   
plt.show()
```

```python
plt.hist(df["age"].dropna(), bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("Age_distribution.png") 
plt.show()
```
