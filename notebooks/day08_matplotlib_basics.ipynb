```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
product_A = [200, 250, 180, 300, 320]
product_B = [150, 230, 210, 280, 310]
product_C = [125, 240, 219, 260, 350]

plt.plot(days, product_A, label="Product A", color='green')
plt.plot(days, product_B, label="Product B", marker='o', linestyle='--', color='yellow')
plt.plot(days, product_C, label="Product C", color='orange')

plt.title("Sales Comparison")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()

plt.savefig("sales_comparison.png")   
plt.show()
```
