```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("titanic")

df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["deck"])
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="survived", hue="survived", data=df)
plt.title("Survival Count")
plt.show()
```

```python
ax = sns.countplot(x="sex", hue="survived", data=df)
plt.title("Survival by Gender")
for container in ax.containers:  #Show Numbers On the Graph
    ax.bar_label(container)
plt.show()
```

```python
pd.crosstab(df["sex"], df["survived"])
```

```python
pd.crosstab(df["pclass"], df["survived"])
```

```python
sns.histplot(data=df , x="age", hue="survived", bins=30, kde=True)
plt.title("Age distribution by Survival")
plt.show()
```

```python
pd.crosstab([df["pclass"], df["sex"]], df["survived"])
```

```python
sns.histplot(df["fare"], bins=30)
plt.title("Fare Distribution")
plt.show()
```

```python
pd.crosstab(df["embarked"], df["survived"])
```
